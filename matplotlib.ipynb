```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5, 11)
```

```python
x
```

```python
y = x*x
```

```python
y
```

```python
#funcional
plt.plot(x,y, 'r--')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.title('Titulo')
```

```python
plt.subplot(1, 2, 1)
plt.plot(x,y, 'r--')
plt.subplot(1,2,2)
plt.plot(x, y, 'g*-')
```

```python
fig= plt.figure()
```

```python
fig= plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.3,0.3])


axes1.plot(x,y)
axes1.set_xlabel('Eixo x')
axes1.set_title('Title')
axes2.plot(x,y)
```

```python
fig, ax =plt.subplots()
ax.plot(x, x**3, 'b--')
ax.plot(x, x**4, 'g-.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Titulo')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
```

```python
ax[0].plot(x, y, 'b')
ax[0].set_title('titulo')

ax[1].plot(x, y, 'b')
ax[1].set_title('titulo')
fig
```

```python
fig, ax = plt.subplots(nrows=5, ncols=5)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(figsize=(8,5))
axes.plot(x,y,'r-', label='x^2')
axes.plot(y,x,'g--', label='x**1/2')
axes.set_title('Titulo')
axes.legend(loc=0)
```
